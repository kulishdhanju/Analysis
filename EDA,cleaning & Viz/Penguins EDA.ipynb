```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pp = sns.load_dataset('penguins')
```

```python
pp
```

```python
pp.info()
```

```python
ppc = pp.dropna()
```

```python
ppc
```

```python
ppc.describe(include='all')
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2,2,1)
sns.boxplot(x= 'bill_length_mm' , data=ppc)

plt.subplot(2,2,2)
sns.boxplot(x= 'bill_depth_mm' , data=ppc)

plt.subplot(2,2,3)
sns.boxplot(x= 'flipper_length_mm' , data=ppc)

plt.subplot(2,2,4)
sns.boxplot(x= 'body_mass_g' , data=ppc)

plt.show()
```
