## Q. What duration is threshold above which we can call it long or short?

#### Let's find minimum of 'long' kind

```python
long_min_dur = geyser[geyser['kind'] == 'long']['duration'].min()
long_min_dur
```

#### Let's find maximum of 'short' kind

```python
short_max_dur = geyser[geyser['kind'] == 'short']['duration'].max()
short_max_dur
```

```python
# Set the figure size
plt.figure(figsize=(20, 10))

# First subplot
ax1 = plt.subplot(2, 1, 1)
sns.boxplot(x='duration', data=geyser[geyser['kind'] == 'long'], ax=ax1)
plt.title('Long kind duration')

# Second subplot
ax2 = plt.subplot(2, 1, 2, sharex=ax1)
sns.boxplot(x='duration', data=geyser[geyser['kind'] == 'short'], ax=ax2)
plt.title('Short kind duration')

# Show the plot
plt.show()
```

### Thus, threshold can be any value between 2.5 to 4.0 aprroximately.

## Let's compare for each mode

### a) Duration

```python
# Set the figure size
plt.figure(figsize=(10, 20))


# First subplot
ax1 = plt.subplot(4, 1, 1)
sns.boxplot(x='duration', data=geyser[geyser['kind'] == 'long'], ax=ax1)
plt.title('Box plot of Long kind duration')

# First subplot
ax2 = plt.subplot(4, 1, 2, sharex = ax1)
sns.histplot(x='duration', data=geyser[geyser['kind'] == 'long'], ax=ax2)
plt.title('Histogram of Long kind duration')

# Second subplot
ax3 = plt.subplot(4, 1, 3, sharex=ax1)
sns.boxplot(x='duration', data=geyser[geyser['kind'] == 'short'], ax=ax3)
plt.title('Box plot of Short kind duration')

# Second subplot
ax4 = plt.subplot(4, 1, 4, sharex=ax1)
sns.histplot(x='duration', data=geyser[geyser['kind'] == 'short'], ax=ax4)
plt.title('Histogram of Short kind duration')

# Show the plot
plt.show()
```

### b) Waiting

```python
# Set the figure size
plt.figure(figsize=(10, 20))

# First subplot
ax1 = plt.subplot(4, 1, 1)
sns.boxplot(x='waiting', data=geyser[geyser['kind'] == 'long'], ax=ax1)
plt.title('Box plot of Long kind duration')

# First subplot
ax2 = plt.subplot(4, 1, 2, sharex = ax1)
sns.histplot(x='waiting', data=geyser[geyser['kind'] == 'long'], ax=ax2)
plt.title('Histogram of Long kind duration')

# Second subplot
ax3 = plt.subplot(4, 1, 3, sharex=ax1)
sns.boxplot(x='waiting', data=geyser[geyser['kind'] == 'short'], ax=ax3)
plt.title('Box plot of Short kind duration')

# Second subplot
ax4 = plt.subplot(4, 1, 4, sharex=ax1)
sns.histplot(x='waiting', data=geyser[geyser['kind'] == 'short'], ax=ax4)
plt.title('Histogram of Short kind duration')

# Show the plot
plt.show()
```
